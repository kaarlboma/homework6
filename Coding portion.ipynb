```python
import numpy as np
from LinearBandits import LinearBandit
import matplotlib.pyplot as plt
from UCB import UCB
```

```python
# Part a
regret_5 = LinearBandit(5, 1, None, 30000, np.random.randn(5), 0.1)
regret_10 = LinearBandit(10, 1, None, 30000, np.random.randn(10), 0.1)

plt.plot(np.cumsum(regret_5), label='d=5')
plt.plot(np.cumsum(regret_10), label='d=10')
plt.xlabel('Time')
plt.ylabel('Cumulative Regret')
plt.title('Cumulative Regret vs Time, n = 30000')
plt.legend()
plt.show()
```

```python
from itertools import product
arms = np.array(list(product([-1, 1], repeat=5)))
theta_star = np.random.randn(5)
regrets_oful = LinearBandit(5, 1, arms, 30000, theta_star, 0.1)
regrets_ucb = UCB(0.1, arms, 30000, theta_star)

plt.plot(np.cumsum(regrets_oful), label='OFUL')
plt.plot(np.cumsum(regrets_ucb), label='UCB')
plt.xlabel('Time')
plt.ylabel('Cumulative Regret')
plt.title('Cumulative Regret vs Time (Finite Action Set), n = 30000')
plt.legend()
plt.show()
```
